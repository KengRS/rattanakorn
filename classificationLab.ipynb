```python
import pandas as pd
```

```python
df=pd.read_csv('fitnessdata.csv')
df.
```

```python
df=df.set_index('customer_id')
df.head()
```

```python
df=df.drop(columns='name')
df.head()
```

## แบ่งแบบ Split Test

```python
from sklearn.model_selection import train_test_split
```

```python
feature=df.iloc[:,0:6]
feature.head()
```

```python
label=df.iloc[:,-1:]
label.head()
```

```python
feature.shape,label.shape
```

```python
feature
```

```python
label
```

```python
train_feature,test_feature,train_label,test_label=train_test_split(feature,label,test_size=0.4,shuffle=False)
```

```python
train_feature.shape,test_feature.shape
```

```python
train_feature
```

```python
test_feature
```

```python
# สร้าง knn model
from sklearn.neighbors import KNeighborsClassifier

# ประเมินความถูกต้องของ model
from sklearn.metrics import accuracy_score

# ไม่ให้แสดงผล warning
import warnings
warnings.filterwarnings('ignore')
```

```python
knnModel=KNeighborsClassifier(n_neighbors=3)
```

```python
knnModel.fit(train_feature,train_label)
```

```python
result=knnModel.predict(test_feature)
result
```

```python
test_label
```

```python
accuracy_score(test_label,result)*100
```
